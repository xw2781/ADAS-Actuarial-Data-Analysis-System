# Functions

```python
from XLToolBox import *
import stat

def check_stat(file_path):
    # print(os.path.basename(file_path))
    if os.stat(file_path).st_mode == 33060:
        print('Locked')
    else:
        print('unlocked')

def extract_vba_from_xlam(xlam_path, output_dir):
    # excel = win32com.client.Dispatch("Excel.Application")
    excel.Visible = False

    addin = Workbook(xlam_path).COM

    for vba_project in addin.VBProject.VBComponents:
        if vba_project.Type == 1:  # vbext_ct_StdModule (Standard Module)
            module_name = vba_project.Name
            module_code = vba_project.CodeModule.Lines(1, vba_project.CodeModule.CountOfLines)
            
            # Save the VBA code to a file
            with open(f"{output_dir}//{module_name} {datetime.now().strftime('%Y-%m-%d %H-%M-%S')}.txt", "w") as file:
                file.write(module_code.replace('\r\n', '\n'))

    # addin.Close(SaveChanges=False)

addin_path_1 = r"E:\ResQ\Excel Add-ins\ResQToolBox-UI.xlam"
addin_path_2 = r"E:\ResQ\Excel Add-ins\ResQToolBox.xlam"

def load_addin_1():
    File(addin_path_1).set_read_only(0)
    excel.Workbooks.Open(addin_path_1)

def load_addin_2():
    File(addin_path_2).set_read_only(0)
    excel.Workbooks.Open(addin_path_2)
    
def unlock():
    File(addin_path_1).set_read_only(0)
    File(addin_path_2).set_read_only(0)

def lock():
    File(addin_path_1).set_read_only(1)
    File(addin_path_2).set_read_only(1)

def auto_lock():
    while True:
        active_list = get_active_workbook_name()
        time.sleep(10)
        if 'ResQToolBox-UI.xlam' not in active_list:
            File(r"E:\ResQ\Excel Add-ins\ResQToolBox-UI.xlam").set_read_only()
        if 'ResQToolBox.xlam' not in active_list:
            File(r"E:\ResQ\Excel Add-ins\ResQToolBox.xlam").set_read_only()
```
